```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
youtube_chart_data = [ pd.read_csv(f"../data/youtube data/data_{i}/Chart data.csv") 
                      for  i in range(1,8)]
combined_youtube_chart_data = pd.concat(youtube_chart_data)
youtube_table_data= pd.read_csv("./youtube_table_cleaned.csv")

complet_youtube_data = pd.merge(combined_youtube_chart_data , 
                                youtube_table_data[['Video title', 
       'Average view duration', 'Stayed to watch (%)', 'Subscribers gained',
       'Likes', 'Dislikes', 'Comments added', 'Engaged views',
       'Watch time (hours)', 'Subscribers', 'Impressions',
       'Impressions click-through rate (%)', 'Caption_Snippet',
       'Day_publish_time', 'Month_publish_time', 'Year_publish_time']] , on='Video title')
complet_youtube_data.head()
```

```python
def youtube_eda(complet_youtube_data):
    print("\n unique values")
    display(complet_youtube_data['Video title'].unique())
    print("\n first 5 rows")
    display(complet_youtube_data.head())
    display(complet_youtube_data.tail())
    print("\n description")
    display(complet_youtube_data.describe())
    print("\n data types ")
    display(complet_youtube_data.dtypes)
    print("\n missing data")
    print(complet_youtube_data.isnull().sum())
youtube_eda(complet_youtube_data)
```

```python

def youtube_cleaning(complet_youtube_data):
    complet_youtube_data.drop('Content' , axis=1 , inplace=True)
    complet_youtube_data = complet_youtube_data[complet_youtube_data['Views'] != 0]
    complet_youtube_data['Date'] = pd.to_datetime(complet_youtube_data['Date'] , errors= 'coerce')
    # complet_youtube_data.drop('Video publish time' , axis=1 , inplace=True)
    # complet_youtube_data['Day_date'] = complet_youtube_data['Date'].dt.day
    # complet_youtube_data['Month_date'] = complet_youtube_data['Date'].dt.month
    # complet_youtube_data['Year_date'] = complet_youtube_data['Date'].dt.year
    complet_youtube_data['Views'] = pd.to_numeric(complet_youtube_data['Views'] , errors = 'coerce')
    return complet_youtube_data

complet_youtube_data = youtube_cleaning(complet_youtube_data)
```
